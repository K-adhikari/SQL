Question 3. Which schools have highest safety score?

```python
%sql SELECT NAME_OF_SCHOOL FROM Chicago_Publicschool_data WHERE \
    SAFETY_SCORE = (SELECT MAX(SAFETY_SCORE) FROM Chicago_publicschool_data)
```

Question 4. What are the top 10 schools with the highest "Average Student Attendance"?

```python
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM Chicago_publicschool_data ORDER BY \
    AVERAGE_STUDENT_ATTENDANCE DESC LIMIT 10
```

Question 5. List 5 schools with lowest average student attendance.

```python
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM Chicago_publicschool_data ORDER BY \
    AVERAGE_STUDENT_ATTENDANCE \
    LIMIT 5
```

Question 6. Remove the '%' sign from the above result.

```python
%sql SELECT NAME_OF_SCHOOL, REPLACE(AVERAGE_STUDENT_ATTENDANCE, '%', '') AS AVG_STUDENT_ATTENDANCE \
    FROM Chicago_publicschool_data ORDER BY AVG_STUDENT_ATTENDANCE LIMIT 5
```

Question 7. Which schools have average student attendance lower than 70%?

```python
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE \
    FROM Chicago_publicschool_data WHERE \
    CAST(REPLACE(AVERAGE_STUDENT_ATTENDANCE, '%', '') AS DOUBLE) < 70 \
    ORDER BY AVERAGE_STUDENT_ATTENDANCE
```

Question 8. Get the total college enrollment for each community area.

```python
%sql SELECT COMMUNITY_AREA_NAME, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLEMENT \
    FROM Chicago_publicschool_data GROUP BY COMMUNITY_AREA_NAME
```

Question 9. Get 5 community areas with the least total college enrollment sorted in ascending order

```python
%sql SELECT COMMUNITY_AREA_NAME, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT \
    FROM Chicago_publicschool_data \
    GROUP BY COMMUNITY_AREA_NAME ORDER BY TOTAL_ENROLLMENT LIMIT 5
```

Question 10. List 5 schools with lowest safety score.

Here, we get a list with safety scire of 'None'. Let's try to get the schools with less safety score values.

```python
%sql SELECT NAME_OF_SCHOOL, SAFETY_SCORE \
    FROM Chicago_publicschool_data WHERE SAFETY_SCORE !='None' \
    ORDER BY SAFETY_SCORE LIMIT 5
```

Question 11. Get the hardship index for the community area of the school which has College Enrollment of 4368.

For this, we need to first load another dataset.

```python
df = pd.read_csv("./ChicagoCensusData.csv")
df.to_sql("CENSUS_DATA", con, if_exists='replace', index=False,method="multi")
```

```python
%%sql 
SELECT hardship_index FROM CENSUS_DATA CD, Chicago_publicschool_data CPS
WHERE CD.community_area_number = CPS.community_area_number AND college_enrollment = 4368
```

Question 12. Get the hardship index for the community area which has the highest value for college enrollment

```python
%%sql
SELECT community_area_number,community_area_name, hardship_index FROM CENSUS_DATA 
    WHERE community_area_number IN  
    (SELECT community_area_number FROM Chicago_publicschool_data ORDER BY college_enrollment DESC LIMIT 1)
```
